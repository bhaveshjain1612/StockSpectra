```python
def retrieve_api(symbol):
    if symbol.upper()[-3:]==".NS":
        symbol = symbol.upper()
    else:
        symbol = symbol.upper()+".NS"
    return(yf.Ticker(symbol))
```

```python
#balance sheet
def get_company_balance_sheet(result):
    balance_sheet = result.balance_sheet
    return(balance_sheet)
```

```python
#cashflow
def get_company_cashflow(result):
    cashflow = result.cashflow
    return(cashflow)
```

```python
#income statement
def get_company_income_stmt(result):
    income_stmt = result.income_stmt
    return(income_stmt)
```

```python
import plotly.graph_objects as go
def piotorski_plot(score):
    fig = go.Figure(go.Indicator(
        mode = "gauge+number",
        value = score,
        domain = {'x': [0, 1], 'y': [0, 1]},
        title = {'text': "Piotorski Score", 'font': {'size': 24}},
        delta = {'reference': 9, 'increasing': {'color': "RebeccaPurple"}},
        gauge = {
            'axis': {'range': [None, 9], 'tickwidth': 1, 'tickcolor': "darkblue"},
            'bar': {'color': "purple"},
            'steps': [
                {'range': [0, 5], 'color': 'red'},
                {'range': [5, 7], 'color': 'yellow'},
                {'range': [7, 9], 'color': 'green'}],
            'threshold': {
                'line': {'color': "purple", 'width': 8},
                'thickness': 1,
                'value': score}}))
    return fig
```

```python
piotorski_plot(4).show()
```

```python
def collect_vals(income_stmt, balance_sheet, cash_flow):
    def get_vals(val,col,data):
        try:
            x = data[col][0]
        except: 
            x = None
        try:
            y = data[col][1]
        except:
            y = None
        return {val+"_current":x,val+"_previous":y}

    #net incomes  - p1
    net_income = get_vals("net_income","Net Income",income_stmt)
    
    revenue = get_vals("revenue","Total Revenue",income_stmt)
    
    total_assets = get_vals("total_assets","Total Assets",balance_sheet)
    
    operating_cash_flow = get_vals("operating_cash_flow","Operating Cash FlowOperating Cash Flow",cash_flow)
    
    long_term_debt = get_vals("long_term_debt","Long Term Debt",balance_sheet)
    
    current_assets = get_vals("current_assets","Current Assets",balance_sheet)
    
    current_liabilities = get_vals("current_liabilities","Current Liabilities",balance_sheet)
    
    gross_profit = get_vals("gross_profit","Gross Profit",income_stmt)
    
    output_dict = {"Net Income":net_income,
                  "Revenue":revenue,
                  "Total Assets":total_assets,
                  "Operating Cash Flow":operating_cash_flow,
                  "Long Term Debt":long_term_debt,
                  "Current Assets":current_assets,
                  "Current Liabilities":current_liabilities}
    
    return output_dict
```

```python
collect_vals(income_stmt, balance_sheet, cash_flow)['Operating Cash Flow']['operating_cash_flow_current'] 
```
